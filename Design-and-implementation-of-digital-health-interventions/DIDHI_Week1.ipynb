### **Human Centered Design in practice**
- There are many common interpretations of design, the most common being making things more aesthetically pleasant. Although aesthetics is a part of it, human-centred design is much more:
  - It can **increase the effectiveness of disease education tools** and **informatics systems**
  - It can lead to **re-thinking of workflow and communication processes**
  - It can **simplify complex problems**
  - It can **reduce friction in a system**
  - It can create **truly engaging solutions**
- Design and designers bring well established processes for **solving challenging systematic problems**

- **Empathy** involves seeing the world through the eyes of the people involved (e.g. patient, clinician, caregiver or other system stakeholder).
- **Design research** is a set of methods that **enables teams to empathise with users**. The goal is to **uncover people's needs, motivations and behaviours**. There are two main methods to do this:
  - **First-hand observations of users** in their daily lives or at work
  - **Interview of users** one-to-one or with a partner/friend
- There is a **scale issue**: shadowing or interviewing hundreds or thousands of people is nearly impossible. However, **design research** is not about trends or the quantification of results, but it is instead about **stories** that allow us to discover people's true lived experiences. For example, we get to know **workarounds** that people use to make tasks easier or about the **emotion** underpinning an experience.
- The best way to capture these stories is to **go to people in their homes or at work**, rather than people coming to you. There is a lot of **tacit information** we can learn about people by visiting them in their own environment. When we bring people to us for a focus group, they end up learning quite a lot more about us than we do about them.
- Design research is far **afield** from traditional approaches to learn about users, such as market research. In market research, surveys and focus groups are methods to identify existing trends about a particular issue. Instead, design research seeks to **uncover the unarticulated needs** that serve as **inspiration for new ideas**.
- Designers look for **stories** and ask **open-ended questions** and then synthesising the observations to **identify latent needs motivations and beahviours**.

- **Prototyping** does not require specific skills. It is instead a **mindset of iteration and sharing**. It involves the **iterative process of making ideas tangible**. In the **design of products or systems**, prototyping enables teams to **test their assumptions and insights**. Prototypes, by definition, are **low commitment** and **low cost**, enable teams to **work quickly and adjust before arriving at a final solution**.
- Prototypes can take any form: from a hand-drawn sketch to a cardboard structure, to a paragraph text describing an idea.
- When building prototypes, the **learning process** is **ongoing** rather than something that is solely goal-oriented. Prototyping is the **start of an agile flexible final solution**.
- **Prototypes** are meant to be **shared with users and stakeholders** for **feedback** and **constant iteration**. One of the biggest pitfalls in prototyping is not sharing early enough. 
- We are not looking for something statistically significant but for the **gut feeling of some extreme users** (i.e. people who sit outside the norm of our general users). By iterating based on feedback from the extremes, you will pull the standard user along you.

### **Professional design**
- A common minsconception about design is that it is just one discipline. In reality, professionally trained designers come from a variety of disciplines and the desing of users experience requires the input of a multitude of professionals:
  - **Visual & graphic design**: the art and practice of planning and projecting ideas and experiences with visual or textual content. The form this can take can be either physical or virtual and include many things, such as images, words or graphics.
  - **Architecture**: the art and technique of designing and building (distinct skills from constructing). It requires a great deal of empathy for end users. Architects need to understand how people use space and ensure that a really large scale investment would make sense.
  - **Information architecture**: the practice of deciding how to arrange the parts of something (e.g. data, text, education information, ...) to be understandable. It helps to guide people to the right information at the right time.
  - **Industrial design**: a process of design applied to products that are to be manufactured. It focusses on the appearance of a product, how it works and the value it provides to its users. Often, ergonomics are considered.
  - **Human factors**: they examines the relationship between human beings and the systems with wich they interact, by focussing on improving efficiency, creativity, productivity and satisfaction. The goal is about minimising errors (this is important also for architecture and industrial design).
  - **Interaction design**: the design of the interaction between users and products. It often involves elements like statics, motion, sound and space. In the design of virtual interfaces, often interaction designers and visual designers work very closely (or are the same person)
  - **Sound design**: it incorporated the use of sound to evoke emotion or action in a product or service. Sound designers are sorely needed in healthcare environments (e.g. to help with alert fatigue)
  - **Human-computer interaction**: a multidisciplinary field of study focussed on the design of computer technology and the interaction between the users and computers

- All these disciplines add up to **User experience design** (or **UX Design**). It encompasses countless design decisions made by a diverse group of designers and stakeholders to provide meaningful and relevant experiences to users. Designers often work together in teams, bringing together the appropriate disciplines for the challenge. The team should also comprise patients and any other user.

### **The Principles for Digital Development**
**Reference**  
- *Principles for Digital Development (2012) Principles for Digital Development. Available at: https://digitalprinciples.org/about/ (Accessed: 28 September 2020).*

**Introduction**  
The **nine principles** to be familiar with are:

1. **Design** with the **user**

2. **Understand** the **existing ecosystem**

3. **Design** for **scale**

4. **Build** for **sustainability**

5. Be **data driven**

6. Use **open standards, open data, open source**, and **open innovation**

7. **Reuse** and **improve**

8. **Address privacy** and **security**

9. Be **collaborative**

These nine principles are listed as key, by the Principles for Digital Development webpage (the reference). You can find there an explanation of how they were developed and by whom. They are also not intended to be static:

*"...the Digital Principles may change as they are updated and modified to fit specific contexts and environments. The Digital Principles community plays a key role in ensuring the Principles are still relevant and useful."*

Visit the reference website for comprehensive information and resources about each of the principles: Each principle is covered by a wealth of advice and information, including core tenets and ways of implementing each principle throughout the stages of the project life cycle from planning to implementation and evaluation.

Here we summarise each of the current principles, and where appropriate, point to content in other parts of this specialisation. 

 1. **Design with the user**  
The **user** should be at the **heart of your design**. The principles suggest you should not just be designing for your users, you should be **designing with them**. **User-centred design** means incorporating them at every stage of the process, from finding their requirements from the start, getting feedback through prototyping, and iteratively building and testing as you develop. Next week's content includes more detailed items on user-centered design, where it is referred to as Human-Centred Design or Design Thinking. See the first lesson in Week 5, "Introduction to Design".

2. **Understand the existing ecosystem**  
Every **context** and **setting** will be different. It is imperative to consider the ecosystem - the **region**, the **community**, that you are working within, with respect to its needs, and what structures already exist. This will help tools you design and develop to be useful, relevant, and sustainable (and not duplicate existing work). Anything which affects individuals' and communities' ability to participate, access and use any tools or initiatives must be taken into account. Things to consider within your ecosystem include:
  - **culture**
  - **gender norms**
  - **political environment** 
  - **economy**
  - **technology infrastructure**  
  You will be introduced to a framework for thinking about your setting in week 8 - the **PESTLE framework**, which lists these aspects to consider.

3. **Design for scale**  
"Scale" might be dependent on context, but it should at least mean "**thinking beyond the pilot phase**". It can be difficult, especially in the field of Digital Health, to transition projects from a pilot stage to more widespread implementation. The Digital Principles website, under this principle, mentions the term "**pilotitis**" or the **inability to move initiatives beyond the pilot stage**, as identified by the **mHealth community** in particular. But you should hope, and therefore plan, for your project to move in this direction. If you do this from the beginning you should be able to more easily expand. Whether this is to a **wider area**, or a **different area** (or market, or region, or country...), if you have **designed for scaling up** from the beginning, this will be all the more feasible. Choices may have to be made which trade between rapidly starting up at a small scale, or being more careful in starting in a way which means you can be more widely applicable in the future. If you look ahead to Week 7, you will learn about a framework for assessing whether a tool could succeed, and perhaps this can help you think about how you can best design your interventions for success and scalability.

4. **Build for sustainability**  
A programme which is built with sustainability in mind will be **more likely to succeed long-term**. Possibly the most obvious aspect of sustainability of a programme is **whether funding can be maintained** (perhaps the project generates its own revenue, or is funded externally), or even whether the programme could survive losses in funding. Many other aspects feed into sustainability; perhaps broadly, you will need to plan to be **adaptable** or to be **resilient to change** - changes in **user needs**, or changes in the **ecosystem**, or changes in the **technologies** you will be using, especially because technology can become outdated and obsolete very fast.    


 5. **Be data driven**  
Throughout this specialisation you will find that data is absolutely fundamental to digital health applications. Projects should be data driven from beginning to end. The considerations around data in Digital Health are summarised in Week 9. Data must be used to inform decisions, if the data are to have an impact.

6. **Use open standards, open data, open source, and open innovation**  
Open approaches are incredibly important for **facilitating access and collaboration**. If software and tools are not open, if they are only available to specific parties, usually those who can pay for them, this is a barrier to collaboration and especially to **access for those in lower-resource settings**. Paying to use licensed tools can be an unwise use of resources, if open alternatives are available. This applies both to you as the developer of a tool, and to your potential users; so your initiative should as far as possible, use and develop open tools. There are many open standards, software tools, and platforms available; perhaps not everything will fit your exact needs and the principle of using open tools may need to be balanced against the need to match this with your use case. Of course where data is concerned, security and privacy must also be paramount. However, there are many resources to help guide you to choose and use open approaches where possible. If tools and innovation are open, and accessible to all, this is important for **transparency**. Ultimately these approaches are good for **increasing inclusion and engagement** - if other players in your ecosystem cannot be a part of understanding what you are doing, you may find that they are less willing to engage.

7. **Reuse and improve**   
You will rarely be required to start something entirely from scratch; there will often be tools and products to reuse and improve upon, and resources which can be reused. In your ecosystem, find what is already available and use and build on it as appropriate. **Learn from what has been tried and tested** - **collaborate** with others to increase your knowledge of other related tools. You may be building something new, but that does not mean that every part of that will be brand-new. Use and build on the resources already existing.

8. **Address privacy and security**  
Principle 5 emphasises data, and principle 6 encourages the use of open tools. Of course, privacy and security must be maintained at the same time as following those two principles. **Sensitive data** must be gathered using **informed consent**, and **stored securely**. A **risk-benefit analysis** may be useful. The **rules of your ecosystem**  (e.g. GDPR) **relating to privacy** must be followed. Keep the best interests of people especially those whose healthcare data is being used, in mind.

9. **Be collaborative**  
This principle works across all the others. From collaborating with your users, to collaborating with others in your ecosystem - you can learn and share knowledge, insights, resources to the benefit of as many stakeholders as possible. The reference states "No single initiative or organization can make it happen alone" - define what you will contribute, and work with others where their expertise complements yours.

## **1.3 Interoperability and International Data Standards**

### **Interoperability**
- Interoperability is about **how data can be shared across health systems**, between healthcare providers, products and EHRs. If you can't send information efficiently, it is wasteful and potentially a risk to patient safety.
- The **Institute of Electrical and Electronic Engineers** (**IEEEs**) defines interoperability as: the **ability of 2 or more systems or components to exchange information and to use the information that has been exchanged**.
- Two important concepts:
  - **Ability of systems to exchange information** (**health information exchange**)
  - The systems need to be **able to use and understand the information** that has been exchanged (i.e. all systems must use/understand a **common language**).
- For interoperability to work you need **clear standard**, i.e. a document **established by consensus** and **approved by a recognised body** that provides **rules or guidelines for common use**. Thus is aimed at o**ptimising the degree of order** in a given situation.
- Also required are **simple interoperability standards** on **how to transport data** (e.g. **vocabularies** and **terminologies** for the wrods and phrases that are used). **Standards** are also required for **structure**, so messages can be read by computers after they have been **broken down into appropriate information chunks**.
- Currently there are **several different and sometimes competing standards**, although there is **no absolute consensus** about which one is to use. Why is it so difficult?
  - The **language people use in different health systems and different health specialites is highly variable**. It is almost as if each system/specialty has its own dialect.
  - **Each technology provider tends to have their own system and infrastructure and their incentives might not be aligned with interoperability** (easy interoperability means people can switch provideers easily)

-  Despite these difficulties, progress has been made towards common standards. **Two existing standards** that are important to be aware of:
  - **Systematised Nomenclature of Medicine Clinical Terms** (**SNOMED CT**): a **structured clinical vocabulary for use in EHRs**, increasingly the most commonly used language. It encovers symptoms, test results, diagnoses, drugs and devices.
  - **Fast Healthcare Interoperability Resources** (**FHIR**): a global **standard for passing healthcare data between systems**. It is increasing becoming the industry standard.

- **Application Programme Interface (API)**: a **structured communication protocol that allows you to make an ask of information from a server**.


### **Introduction to International Data Standards**
**Data Standards**  
In the video about interoperability, you heard about international data standards. **Standardising data makes combining data from different sources possible**, for analysis and presentation. Standards overcome the hurdles posed by the nature of the data, especially data which you might need for a Digital Health application. Data could be gathered from different sources, recorded by different people and systems, and originating from different regions around the world. Your application must be able to make sense of all the data, and record, combine, and analyse it reliably.

Even when there already seem to be standard terminologies for items, **entering and reading data digitally** requires **entries of variables to be completely standardised**. Humans reading data might understand that words spelled slightly differently, are meant to be the same. You, as a human, might be able to interpret typographical errors easily. Beyond that, you might know that "150mg paracetamol" and "acetaminophen, 150 mg" are actually the same thing. However, if you need to search a large number of records for "paracetamol 150mg", you need to know how many different ways that data might have been entered - or at least, your search system needs to be able to know to search for all the equivalent alternatives. Ideally, you would have a system that means that "paracetamol 150mg" is always entered the same way in any record.

So there needs to be a **set of rules for recording data**. This is where standards come in. In the video, SNOMED is mentioned. The SNOMED website states: **SNOMED CT [Clinical Terms] expressions are used in Electronic Health Records (EHRs) to represent clinical meanings in a way that enables automatic interpretation**.  The automatic interpretation part is key in the world of Digital Health, but standards are important beyond that to mitigate ambiguity and confusion when interpreting data.

**Controlled Terminology**  
Standards are built on controlled terminology, or **structured clinical vocabulary**, which explains how to record items, or what to transform them into when they have already been collected. One international standard, **Clinical Data Interchange Standards Consortium**, **CDISC**, has this to say: 

*"Controlled Terminology does not tell you WHAT to collect; it tells you IF  you collected a particular data item, how you should submit it in your electronic dataset."*

Having controlled terminology means that you always know how to record a particular item (or what to transform it to when you are curating data which was not standardised), so it will be **stored consistently and reliably**, and can therefore be **analysed reliably** as well. Controlled terminology can also help ensure data is **precise** and **recorded at the right level of detail**.

**SNOMED CT**  
Have a look at the reference, *SNOMED International (2020) - the "Five-step briefing"*. This explains fundamentally what SNOMED CT is.

There are comprehensive lists of controlled terminology that you can browse via the SNOMED website. If you search for a particular term in the controlled terminology, for example "paracetamol", you can find all the Controlled Terminology which includes the word "paracetamol". There are many standard doses of many medications, and so all those which are in use are standardised in the SNOMED controlled terminology. The NHS SNOMED CT browser (SNOMED International 2017) shows you how each term is strictly defined, and how each term is related to others. SNOMED CT even includes synonyms, so data entry can be done using terms which might be more familiar according to different users' contexts. 

**International interoperability**  
Now that data can be so easily shared globally, it is more important than ever to have international standards. Furthermore, there are many different potential data sources for Digital Health applications. Using a comprehensive, international standard **facilitates the collation and use of data from many countries, many different health systems, and many different data sources** such as patient records or data from wearables. Data can be mapped from one standard to another where necessary.

## **1.4: Digital Delivery and Agile Methodologies**

### **Introduction to Digital Delivery**

- **Agile approach to delivery**: it is different from traditional project management, which focusses on pre-planned activities and timelines (waterfall management). Agile is the **most common delivery method for digital and software projects**. In contrast with rigid waterfall delivery, agile delivery is **flexible** and **iterative**. Core principles:
  - **Focus on user needs** in designing the approach to the service or product. Provide maximum benefit to the users (what they want and what they actually use).
  - **Work iteratively**: it is about creating ideas/hypotheses and testing them, learning and improving from the process. A central idea is the creation of a **minimum viable product**, an initial version of the product that can be improved by seeing it used in the real world. The minimum product should be shipped or delivered early in the development process.
  - Agile processes are allowed to **fail fast** and **learn quickly**.
  - Work should be **collaborative** and **non hierarchical**. There should be trust and motivation across the team.
  - Agile work should have a **culture** of **continuous improvement**. There should be regular reflection on what has gone well, what has not and how to become more effective.

- There are **different Agile methodologies**, e.g.: 
  - **Scrum**: people work in short bursts of activity (**sprints**) and focus on delivering particular tasks. There is a retrospective reflection at the end of each sprint to see how things could be improved. There is a focus on meeting user requirements and an understanding these may well change over time.
  - **Kanban**

- In these process, specific people have **specific project roles**:
  - **User researcher**
  - **Content designer**
  - **Software developer**
  - **Product owner** (coordination role)

- **Agile for healthcare**. Agile methodologies emerged in the software industry and there are extra things to consider if applying them in health:
  - **Risk**: the common mantra in technology "*fail fast and fail often*" is less acceptable in health (where the accepted philosophy is instead **evidence based**, characterised by the mantra "*first do no harm*").
  - Agile approaches need to reflect the **multidisciplinary team in health**: it is required by the Agile team to integrate **clinical advice**, **information governance**, **ethics** and **patient perspective**, on top of the conventional delivery disciplines of design and development.

- **Stages of digital delivery**. Agile follows 4 main phases:  
  1) **Discovery**: exploring the problem space. You should try and understand the problem you are trying to solve (e.g. users, constraints and opportunities available)   
  2) **Alpha**: testing options with hypotheses (build initial prototypes and test different ideas)   
  3) **Beta**: building and refeining options (the best ideas from alpha phase are further developed in the real world; also important to consider how the service will integrate with/replace existing services and preparing for the transition to live)   
  4) **Live**: continuously improving (live service).

- **Digital delivery** is about working in a **flexible, adaptive and iterative way**. It involves **understanding and fulfilling users' needs** and requires a **learning culture**.

### **Use of Agile methodologies in relation to Digital Health**

While Agile accommodates past failure to learn and improve on processes and outcomes, in health care, this must be balanced with risk assessments to ensure that no harm befalls patients. In health care, non-hierarchical teams must be multidisciplinary to ensure patients get the best care possible. You’ll see that there is a risk for managers to manage in ways that run counter to Agile principles and practices, undermining the effectiveness of teams. This reading discusses **crucial practices for capitalizing on Agile’s potential**.  

**Summary**  
**Agile innovation methods** have greatly increased success rates in software development, **improved quality** and **speed to market**, and **boosted the motivation and productivity of IT teams**. These methods are spreading across a broad range of industries and functions, not least DH. However, broadly (and in DH in particular), teams do not understand how to promote and benefit from Agile. The authors have discerned s**ix crucial practices for capitalising on Agile’s potential**: 

1. Learn how Agile really works
2. Understand when it is appropriate
3. Start small and let passionate evangelists spread the word
4. Allow teams that have mastered the process to customise their practices
5. Practice Agile at the top
6. Destroy corporate barriers to Agile behaviours.   


These practices are expanded to described Agile approaches like scrum and Kanban. As multidisciplinary teams are essential in DH, the suggested **techniques to destroy barriers to Agile behaviour across teams** are very important. These techniques include: 

- Get everyone on the same page
- Don’t change structures right away; change roles instead
- Name only one boss for each decision
- Focus on teams, not individuals
- Lead with questions, not orders.
