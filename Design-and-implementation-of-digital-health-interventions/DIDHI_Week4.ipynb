## **4.1: Introduction to contextual analysis**

### **An introduction to Contextual Analysis (PESTLE and SWOT)**
  

- You have previously looked at the challenges to digitalisation in healthcare. Now, you are going to look at how you can work to overcome these challenges, through **designing a robust and effective strategy to support implementation** of your idea. After you learn the basics of strategy development, you will have the opportunity to look at some examples of successful digital health strategy. 

- Broadly speaking, a **strategy is a plan for achieving a desired result** (1). This often comprises a set of actions or steps that you want to take to solve a problem or make a change. In reality, the process of developing a strategy encompasses much more than writing out a list of actions: it is essential to tailor your plans to the specific situation at hand, and ensure you know how you will implement them. To help you do this, you may choose to use a **strategic development model**, or employ a strategic planning process.   

- There are many **frameworks for strategic development**, but most models contain **four basic elements** (2): 
  - **Contextual analysis**
  - **Strategy formulation**
  - **Strategy implementation**
  - **Monitoring and evaluation**   

- Today, you are going to focus on the **contextual analysis aspect of strategic planning**, that is to say **understanding the wider world in which your project or organisation sits**. This includes describing both the **internal** (within your organisation or team) and **external** (outside of your organisation or team) environment, as well as considering the implications of these environments for achieving your overall aim(3).

- When undertaking **contextual analysis**, it can be useful to make use of **frameworks** or **checklists**, which prompt you to consider all the relevant features of the environment. Two such frameworks are **PESTLE** and **SWOT**, both of which are checklists named using acronyms for their component parts. While **PESTLE** is mainly focused on the **external environment**, **SWOT** invites consideration of **both internal and external factors**.

#### **PESTLE**

- PESTLE is a useful tool for considering the **wider macro environment**, and documenting **interdependence** and **complexity** (1). Each of the letters in PESTLE stands for a type of factor that exists external to your organisation, but is capable of influencing what happens inside it (1,5):

- To use PESTLE, you should consider how the above factors may help or hinder your organisation (1). While this may start as a list of influences, it is crucial to move on to analyse these influences, by considering how they will impact your organisation and how you could respond to them (1).   

- The original acronym for this tool was simply **PEST** (Political, Economic, Sociological, Technical), with Legislative and Economic factors added at a later stage. Today, various iterations of the tool exist, including **PESTLE**, **PESTELI** (with the addition of Industry)(1), and **DEPESTELI** (with the addition of Demographic and Epidemiological) (5). You can use whichever version suits your project best.

#### **SWOT**

- SWOT stands for **Strengths, Weaknesses, Opportunities, and Threats**. This tool considers **both features internal to an organisation** (its **strengths** and **weaknesses**) and **external** to it (any **opportunities** and **threats**). It was developed in the 1960s, and remains one of the most frequently used frameworks in strategic planning today (1,6).

- SWOT can be applied across a wide range of industries, including digital health(1). It is used to consider why change might be needed within an organisation, and what external factors may support or prevent this. To use the SWOT framework, begin by considering the overall aim of an organisation or project. Once this has been identified, identify the strengths and weaknesses of the project. It can be useful to use a checklist, such as the **McKinsey 7S Model** (**Strategy, Structure, Systems, Staff, Style, Shared Values, Skills**), to ensure you cover all aspects of the organisation (1). To be a true strength, the characteristic must work to enable the organisation to achieve its aim. To be a true weakness, the characteristic must work to hinder the organisation from achieving its aim (3). You should then consider what the causes and consequences of each strength and weakness are. 

- Next, you should turn your gaze outward, to **external opportunities and threats**. It may be useful to use your **PESTLE checklist to support this element of your SWOT analysis** – it will help you conduct a more thorough assessment. Once you have identified the opportunities and threats to your project or organisation, you should consider the impact of each on your project or aim. Opportunities can only be described as such if they will enable achievement of your goal – likewise a threat is only a threat if it impedes it. Once true threats and opportunities have been identified, you should consider what you can do to take advantage of the opportunities, and mitigate the impact of threats (7).

- You can use a **grid or matrix** to support your SWOT analysis, if you find it useful. To do this, construct a two-by-two table that classifies influences as either positive or negative, and internal or external(7). This will give you cells in which you can enter your strengths (positive internal forces), weaknesses (negative internal forces), opportunities (positive external forces) or threats (negative external forces).

image.png

- Both PESTLE and SWOT can help you with your contextual analysis by providing a comprehensive starting point and a checklist to ensure that you consider a wide range of factors. It is important to remember that the aim of these processes is not just to generate a list of factors which might influence the goals of your project, but to consider what the impact might be, and to devise strategies to respond to it. When conducting contextual analyses, it is important to consider both internal and external factors, and using a combination of tools, rather than relying too heavily on one, can help you achieve this. Once you have a good understanding of where your project fits in the wider world, you can move on to formulating, implementing, and evaluating your strategy.

## **4.2: Examples of Digital Health Strategies**

### **Simple Strategy for Digital Health**

- We introduce a couple of **useful frameworks** to help **thinking strategically** about the world around you and how any digital products you develop fit within that. 
- Understanding the **wider context** in which your digital work sits is vitally important. It is not juts about your idea, but the whole system and the culture you operate in that defines digital adoption: **a complex mix of people, culture, organisation and technology**.
- **SWOT analysis**: a **simple structured thinking tool** that asks you to look in turn at your idea through 4 separate lenses (you need to enhance the positives and try to mitigate the negatives):
  - **Strengths**: what is your idea good at and that other ideas can't do? What are the selling points that you can persuade others with?
  - **Weaknesses**: what doesn't your idea do? What are the potential risks of using it?
  - **Opportunities**: think about the advantages you have in the space you have to operate within. What allies are there to support your idea?
  - **Threats**: what barriers might be there to the adoption of your idea? are there competitors? are there cultural barriers?
- The first two concepts of SWOT are largely internal, intrinsically focussed on your product/idea. The last two relate to how your idea might fair in a competitive environment.

- **PESTLE**: it allows you to look in a systematic way to the situation you find yourself operating in. This framework is particularly useful for **looking at the wider macro environment** that exists around you:
  - **P = Political**: the political happenings of the world around (external or internal politics within the organisation)
  - **E = Economic**: the economic and financial circumstances you find yourself in. You need to demonstrate value for money or return on investment.
  - **S = Societal**: what's going on in society and human life around us.This might be questions of acceptability or culture. With modern technology is not only the ability to use it but also its level of acceptance.
  - **T = Technological**: the technology you are using and how it works in the wider technology landscape. How it interact with other technologies, what are its standards and interoperability.
  - **L = Legislative**: it relates specifically to the legal system you are operating in. For digital, this might mean understanding the regulatory or information governance requirements (e.g. GDPR or HIPAA in US)
  - **E = Environmental**: think about environmental and sustainability issues.

- These are some of the core frameworks used in **business strategy**. Finding a way to align these business frameworks with traditional thinking of public health and evidence-based medicine is part of the challenges of digital health.

### **SWOT analysis of the IT and healthcare industry**
- In the earlier activity, you were introduced to Simple Strategy and one aspect that you learnt about was SWOT analysis. To contextualise this area further, we would like you to read this article by Helms et al. on Information technology and the healthcare industry: A SWOT analysis.

**Reference**

- Helms, M. M. , Moore, R. & Ahmadi, M. (2008) Information Technology (IT) and the Healthcare Industry: A SWOT Analysis. International Journal of Healthcare Information Systems and Informatics (IJHISI). 3(1):75-92. Available from: https://www.researchgate.net/publication/220168683_Information_Technology_IT_and_the_Healthcare_Industry_A_SWOT_Analysis

**Summary**

- This article by Helms et al. describes a **Strength, Weakness, Opportunity and Threat analysis (SWOT)** for Information Technology (IT) and the healthcare industry. 

- The article starts by assessing the **challenges faced within the healthcare industry** including, but not limited to, high cost of operations, inefficiency, inadequate safety, insufficient access to information and poor financial performance, analysts, practitioners, patients and others have concerns for healthcare industries. Even though heaLth industries invest far less in IT (which could also be considered as Digital Health Technologies), IT has been identified as a catalyst to help address these concerns for healthcare. IT systems promote quality, safety, and efficiency, reduce operating cost and reduce medical errors. The article then focuses on the SWOT methodology itself (the popular strategic analysis tool that has been introduced in the video), and its key features.

- The focus of the article then turns to **internal strengths** supporting further industry investment in IT to include improved patient safety, greater operational efficiency, and current investments in IT infrastructure. The strategic initiative to increase the role of IT in healthcare can advance the cause of greater patient safety by enhancing the quality of care. IT also improves operational efficiency and increases productivity by reducing paperwork, automating routine processes, and eliminating waste and duplication. Organisations are improving existing IT infrastructure and personnel to meet growing demands to achieve more efficient operations and more effective levels of healthcare. 

- The **internal weaknesses** include a lack of information system integration, user resistance to new technologies and processes, and slow adoption of IT. The coordination and integration of internal data improves operations and decision making. There is the need to promote user acceptance and organisational adaption of IT systems in Healthcare. 

- The **external opportunities** for the use of IT in healthcare are then examined. This includes the internet, favourable external environment and industry standards. There is a need to further incorporate the internet in healthcare operations to allow new ways of communication between the public, patients and physicians.  To maximise the benefits of IT globally, there should be worldwide support for the utilisation of more IT in healthcare. Globally, standardisation in healthcare IT systems can result in greater levels of system integration, increase data sharing and more powerful data mining. 

- The **external threats** include legal compliance, loss of patient trust and costs. There are risks to patient privacy and safety therefore legal instruments like The Health Insurance Portability and Account ability Act (HIPAA) must be complied with. Strong legal instruments and patient involvement maximise patient safety and retain patient trust. Wide adaptation of IT is expensive. However, if implemented efficiently, IT adoption is consistently related to improved financial outcomes both overall and operationally. 

- This SWOT analysis demonstrates how IT could ultimately improve healthcare but first, it would require the engagement of all stakeholders. The more stakeholders (e.g. patients, physicians, insurers, pharmaceutical industry etc.)  who are involved in IT selection and implementation planning, the greater their acceptance and rate of adoption would be.
